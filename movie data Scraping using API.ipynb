```python
import requests
import pandas as pd
```

```python
response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=046c0cef9b8b630a89195a90cfaa61c6&language=en-US&page=1').json()
```

```python
original_language = []
overview=[]
popularity= []
release_date=[]
title=[]
vote_count=[]
vote_average=[]



for i in response['results']:
    original_language.append(i['original_language'])
    overview.append(i['overview'])
    popularity.append(i['popularity'])

    

    title.append(i['title'])
    vote_count.append(i['vote_count'])
    vote_average.append(i['vote_average'])
    
d={"title":title,"overview":overview, "popularity":popularity,"vote_count":vote_count, "original_language":original_language,"vote_average":vote_average}

df=pd.DataFrame(d)
    
df

    
```
